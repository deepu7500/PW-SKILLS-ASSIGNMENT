```python
ans 2->
Scatter plots are the graphs that present the relationship between two 
    variables in a data-set. It represents data points on a two-dimensional plane or on a Cartesian system
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("this is my x-axis")
plt.ylabel("this is my y-axis")
plt.title("this is a scatter plot")
```

```python
ans 3->  subplot() function adds subplot to a current figure at the specified grid position. 
It is similar to the subplots() function however unlike subplots() it adds one subplot at a time. 
So to create multiple plots you will need several lines of code with the subplot() function.
```

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = [0, 1, 2, 3, 4, 5]
y1 = [0, 100, 200, 300, 400, 500]
x2 =[0, 1, 2, 3, 4, 5]
y2 =[50, 20, 40, 20, 60, 70]
x3 = [0, 1, 2, 3, 4, 5]
y3 = [10, 20, 30, 40, 50, 60]
x4 = [0, 1, 2, 3, 4, 5]
y4 = [200, 350, 250, 550, 450, 150]
fig, axs = plt.subplots(2, 2)
fig.suptitle('Four Line Plots')
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')
plt.show()
```

ans 5->
A box plot is a graphical way to show the distribution of a numeric variable using boxes and lines.
Box plots are useful for comparing groups of data and seeing how they differ in terms of center, spread, and shape.
Box plots are commonly used in many fields, including statistics, finance, biology, and social sciences. 


```python
import numpy as np
import matplotlib.pyplot as plt
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1,box2])
plt.xlabel("box1")
plt.ylabel("box2")
plt.title("Box Plot ")

plt.show()
```
