## ans 1->
Data encoding is the process of converting data from one representation to another, usually to facilitate its
  use in analysis or machine learning.The purpose of data encoding is to represent data in a way that can be easily processed
    and analyzed by algorithms.
    
Data encoding useful to covert  categorical data into  numerical data which is useful in data science to analyse data.
Data encoding help in make structured  data from unstructured data.

# ans 2->
nominal encdoing is also known as one hot encoding.it is used to convert categorical data into numerical data.
In nominal encoding, each unique category or value in a categorical feature is represented as a binary vector. 

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
import seaborn as sns
df=sns.load_dataset('iris')
df
```

```python
encoded=OneHotEncoder().fit_transform(df[['species']])
encoded.toarray()
```
