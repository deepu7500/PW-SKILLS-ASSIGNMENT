ans 1-> F ive plots that we use using seaborn library
(1) Scatter plot - to explore the relationship between two variables

```python
import seaborn as sns
data=sns.load_dataset("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic")
data
```

```python
sns.scatterplot(x=data.sex,y=data.fare)
```

(2)Displot- displot represents the overall distribution of continuos data variable

```python
sns.displot(data['age'])
```

(3)Relypolt- relplot allows us to visualise how variables within a dataset relate to each othe

```python
sns.relplot(x=data.age,y=data.fare,data=data,hue='class')
```

(4)catplot- it is used to work on categorical data and involve categorical variable.

```python
sns.catplot(x='sex',y='age',data=data)
```

(5)jointplot - The function jointplot() in Seaborn library creates a scatter plot with two histograms at the top and right margins of the graph

```python
sns.jointplot(x=data.age,y=data.fare)
```
