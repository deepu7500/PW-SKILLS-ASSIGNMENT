Imbalanced data refers to a dataset where the distribution of classes or categories is not equal.
For example, in a binary classification problem, if the positive class accounts for only 10% of the total observations,
and the negative class accounts for 90%, then the dataset is imbalanced.
If imbalanced data is not handled, it can lead to biased or inaccurate results in the analysis or modeling

ans 4->Up-sampling involves increasing the number of instances in the minority class to balance the dataset.
The goal of up-sampling is to ensure that the number of instances in the minority class is similar to the number
of instances in the majority class.
For example, in audio compression, a high-quality audio file may be down-sampled to reduce its size while
still maintaining a reasonable level of quality.

Down-sampling, on the other hand, involves reducing the number of instances in the majority class to balance the dataset.
 For example, in audio processing, up-sampling can be used to increase the sample rate of an audio file before applying 
     a digital filter.

ans 5-> Data augmentation is a technique of artificially increasing the training set by creating modified
copies of a dataset using existing data.Some common data augmentation techniques include flipping, rotating, cropping, scaling. 

SMOTE-   Synthetic Minority Oversampling Technique (SMOTE) is a statistical technique for increasing the number of
cases in your dataset in a balanced way. It is a well-known preprocessing approach for handling imbalanced datasets.

ans 6->
Outliers in a dataset are data points that are significantly different from the majority of the data points in the dataset.
Outliers can be caused by errors in data collection, measurement errors, or they may represent true but rare occurrences.

example - [1,2,3,4,6,8,1000,1200].here 1000,1200 are outliers.
It is essential to handle outliers because they can significantly affect the results of data analysis and modeling.
Ignoring outliers can lead to inaccurate conclusions and models that do not perform well on new data.
