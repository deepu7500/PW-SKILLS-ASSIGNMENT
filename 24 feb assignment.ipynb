ans 1-> Here are five common functions of the pandas library with examples:

(1)read_csv() - This function is used to read CSV (comma-separated values) files and create a pandas DataFrame.

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df
```

(2)head() - This function is used to display the first n rows of a DataFrame.

```python
df.head()
```

(3)describe() - This function is used to generate descriptive statistics for a DataFrame

```python
df.describe()
```

(4)groupby() - This function is used to group rows of a DataFrame based on one or more columns.

```python
s=df.groupby("Age")
```

```python
s.sum()
```

```python
s.mean()['Pclass']
```

(4)plot() - This function is used to create various types of plots using the data in a DataFrame.

```python
df['Age'].plot()
```
